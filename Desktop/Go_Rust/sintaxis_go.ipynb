# Introducción a Go (Golang)
Go es un lenguaje de programación moderno, compilado y fuertemente tipado, diseñado para ser eficiente y fácil de usar. Su principal fortaleza es la concurrencia, la rapidez de compilación y la claridad en la sintaxis.
## Usos comunes de Go
- **Desarrollo de APIs y servidores web**: Gracias a su rendimiento y librerías estándar.
- **Microservicios**: Por su facilidad de despliegue y binarios independientes.
- **Herramientas CLI**: Go permite crear ejecutables portables y rápidos.
- **Sistemas distribuidos y cloud**: Proyectos como Docker y Kubernetes están escritos en Go.
- **Automatización y scripts de infraestructura**.
## Gestión de dependencias: Go Modules
Go utiliza un sistema de módulos integrado para manejar dependencias. El comando `go mod init` crea el archivo `go.mod` que registra los paquetes requeridos.
> Los paquetes en Go se agrupan en módulos y se gestionan sin herramientas externas.
## Comandos esenciales
- `go mod init <nombre>`: Inicializa el módulo.
- `go build`: Compila el proyecto.
- `go run archivo.go`: Ejecuta el archivo fuente.
- `go get <paquete>`: Añade una dependencia.
- `go mod tidy`: Limpia y sincroniza dependencias.
- `go test`: Corre los tests.
- `go fmt`: Formatea el código.
- `go vet`: Busca errores comunes.
# Fundamentos de la sintaxis en Go

### Declaración de variables y constantes
Las variables pueden declararse con `var` o usando `:=` dentro de funciones. Las constantes se definen con `const` y pueden usar `iota` para enumeraciones.

```python
import "fmt"

var saludo string = "Hola Mundo"
edad := 20
const PI = 3.14
const (
    Domingo = iota
    Lunes
    Martes
)
fmt.Println(saludo, edad, PI, Lunes)
```

### Tipos de datos básicos y compuestos
Go soporta enteros, flotantes, booleanos, cadenas, arrays, slices, maps, structs y punteros.

```python
var numero int = -7
activo := false
notas := []float64{7.5, 8.0, 9.2}
colores := map[string]string{"rojo": "#FF0000"}
type Libro struct { Titulo string; Paginas int }
libro := Libro{Titulo: "Go Básico", Paginas: 120}
fmt.Println(numero, activo, notas, colores, libro)
```

### Funciones
Las funciones pueden retornar uno o varios valores. Se pueden usar parámetros nombrados y funciones variádicas.

```python
func suma(a, b int) int { return a + b }
func dividir(a, b float64) (resultado float64, err error) {
    if b == 0 {
        err = fmt.Errorf("división por cero")
        return
    }
    resultado = a / b
    return
}
total := suma(2, 3)
res, err := dividir(10, 0)
if err != nil { fmt.Println("Error:", err) } else { fmt.Println(res) }
```

### Condicionales y bucles
Go utiliza `if`, `else`, `switch` y un único bucle `for` que puede comportarse como bucle clásico, while o infinito.

```python
x := 15
if x > 10 {
    fmt.Println("Mayor que 10")
} else {
    fmt.Println("10 o menor")
}
for i := 0; i < 3; i++ { fmt.Println(i) }
for _, v := range []string{"a", "b"} { fmt.Println(v) }
```

### Punteros y paso por referencia
Go permite manipular direcciones de memoria con punteros, pero no permite aritmética de punteros.

```python
y := 5
p := &y
*p = 10
fmt.Println(y) // 10
```

### Structs, métodos e interfaces
Los structs agrupan datos, los métodos se asocian a structs y las interfaces definen comportamientos.

```python
type Animal struct { Nombre string }
func (a Animal) Hablar() string { return "Hola, soy " + a.Nombre }
type Hablador interface { Hablar() string }
perro := Animal{Nombre: "Bobby"}
fmt.Println(perro.Hablar())
```

### Manejo de errores
En Go, los errores se manejan retornando un valor de tipo `error` y comprobando si es distinto de `nil`.

```python
import "errors"
func raizCuadrada(x float64) (float64, error) {
    if x < 0 {
        return 0, errors.New("no se puede raíz negativa")
    }
    return x * x, nil
}
res, err := raizCuadrada(-1)
if err != nil { fmt.Println("Error:", err) }
```

### Panic, recover y defer
`panic` detiene la ejecución, `recover` la captura y `defer` ejecuta código al salir de una función.

```python
func seguro() {
    defer func() {
        if r := recover(); r != nil {
            fmt.Println("Recuperado de:", r)
        }
    }()
    panic("error fatal")
}
seguro()
```

### Composición y polimorfismo
Go utiliza composición de structs y polimorfismo mediante interfaces para reutilizar y extender comportamientos.

```python
type Vehiculo struct { Marca string }
type Auto struct { Vehiculo; Puertas int }
func (a Auto) Info() string { return a.Marca }
miAuto := Auto{Vehiculo: Vehiculo{Marca: "Ford"}, Puertas: 4}
fmt.Println(miAuto.Info())
```
