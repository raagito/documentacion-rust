# Conceptos Fundamentales de Rust

Rust es un lenguaje de programación moderno que prioriza la seguridad, el rendimiento y la concurrencia. Su sistema de manejo de memoria evita errores comunes y permite desarrollar aplicaciones robustas y eficientes.

## ¿Por qué elegir Rust?

- **Seguridad en memoria**: El compilador previene errores como desbordamientos y uso de memoria no válida.
- **Alto rendimiento**: Comparable a C/C++, ideal para sistemas críticos.
- **Concurrencia sin miedo**: Permite escribir código concurrente sin riesgos de condiciones de carrera.
- **Ecosistema moderno**: Herramientas como Cargo facilitan la gestión de dependencias y la construcción de proyectos.

## Convenciones y Estilo
- Variables y funciones: `snake_case`
- Tipos y structs: `PascalCase`
- Constantes: `SCREAMING_SNAKE_CASE`
- Paquetes (crates): `snake_case`

Ejemplo:
```rust
let mi_variable = 10;
fn calcular_area() {}
struct UsuarioActivo {}
const MAX_INTENTOS: u8 = 5;
```

## Herramienta principal: Cargo
Cargo es el gestor de proyectos y dependencias en Rust. Permite compilar, probar y administrar librerías fácilmente.

Comandos útiles:
- `cargo new proyecto` — Crea un nuevo proyecto
- `cargo build` — Compila el proyecto
- `cargo run` — Compila y ejecuta
- `cargo test` — Ejecuta pruebas
- `cargo fmt` — Formatea el código
- `cargo add crate` — Agrega una dependencia
- `cargo update` — Actualiza dependencias

# Sintaxis Básica

## Declaración de Variables
Por defecto, las variables en Rust son inmutables. Para hacerlas mutables, se usa `mut`.

````python
let ciudad = "Madrid";
// ciudad = "Barcelona"; // Error: inmutable
let mut contador = 1;
contador += 1; // Ahora sí es posible
````

### Shadowing
Puedes redeclarar una variable con el mismo nombre para cambiar su tipo o valor.

````python
let dato = 7;
let dato = dato * 2;
let dato = "ahora soy texto";
````

### Constantes
Las constantes requieren tipo explícito y no pueden cambiar ni ser shadowed.

````python
const PI: f32 = 3.1416;
const MENSAJE: &str = "Hola mundo";
````

## Tipos de Datos
- **Escalares**: enteros (`i32`, `u8`), flotantes (`f64`), booleanos (`bool`), caracteres (`char`).
- **Compuestos**: tuplas, arrays, strings.

````python
let activo: bool = false;
let letra: char = 'R';
let tupla: (i32, &str) = (42, "respuesta");
let numeros: [i32; 3] = [10, 20, 30];
let saludo = String::from("¡Hola!");
````

## Expresiones y Declaraciones
Las expresiones devuelven un valor, las declaraciones realizan una acción.

````python
let suma = 3 + 4; // declaración con expresión
let resultado = {
    let x = 2;
    x * 5
}; // El bloque es una expresión
````

## Funciones
Se definen con `fn`. El tipo de retorno se indica con `->`.

````python
fn cuadrado(n: i32) -> i32 {
    n * n
}
fn imprimir_mensaje(msg: &str) {
    println!("Mensaje: {}", msg);
}
````

## Comentarios
- Línea: `// comentario`
- Bloque: `/* ... */`
- Documentación: `///` antes de funciones o structs

````python
// Esto es un comentario de línea
/*
Comentario de bloque
*/
/// Comentario de documentación
fn ejemplo() {}
````

## Condicionales
El `if` no requiere paréntesis. También puede usarse como expresión.

````python
let edad = 20;
if edad >= 18 {
    println!("Mayor de edad");
} else {
    println!("Menor de edad");
}
let estado = if edad > 30 { "adulto" } else { "joven" };
````

### Match
Permite comparar múltiples patrones, similar a switch.

````python
let dia = 3;
let nombre_dia = match dia {
    1 => "Lunes",
    2 => "Martes",
    3 => "Miércoles",
    _ => "Otro día",
};
````

## Bucles
Rust ofrece `loop`, `while` y `for`.

````python
let mut suma = 0;
for n in 1..=5 {
    suma += n;
}
let mut cuenta = 3;
while cuenta > 0 {
    cuenta -= 1;
}
let mut i = 0;
loop {
    if i >= 3 { break; }
    i += 1;
}
````

## Ownership y Borrowing
Rust gestiona la memoria mediante el sistema de propiedad.
- Cada valor tiene un único propietario.
- Al asignar un valor a otra variable, la propiedad se transfiere (move).
- Para compartir sin transferir, se usan referencias (`&`).

````python
let texto = String::from("Hola");
let copia = texto; // texto ya no es válido
let nombre = String::from("Ana");
let longitud = calcular_longitud(&nombre); // nombre sigue siendo válido
fn calcular_longitud(s: &String) -> usize {
    s.len()
}
````

## Structs y Enums
- **Structs**: Agrupan datos relacionados.
- **Enums**: Representan múltiples variantes posibles.

````python
struct Punto { x: i32, y: i32 }
let p = Punto { x: 1, y: 2 };
enum Direccion { Norte, Sur, Este, Oeste }
let dir = Direccion::Norte;
````

## Manejo de Errores
Rust no usa excepciones, sino los tipos `Result` y `Option`.

````python
fn dividir(a: i32, b: i32) -> Result<i32, String> {
    if b == 0 {
        Err("División por cero".to_string())
    } else {
        Ok(a / b)
    }
}
let res = dividir(10, 2);
match res {
    Ok(val) => println!("Resultado: {}", val),
    Err(e) => println!("Error: {}", e),
}
````

## Orientación a Objetos en Rust
Rust no tiene clases, pero puedes usar structs, métodos (`impl`) y traits para lograr encapsulamiento y polimorfismo.

````python
struct Circulo { radio: f64 }
impl Circulo {
    fn area(&self) -> f64 {
        3.1416 * self.radio * self.radio
    }
}
trait Dibujable {
    fn dibujar(&self);
}
impl Dibujable for Circulo {
    fn dibujar(&self) {
        println!("Dibujando un círculo de radio {}", self.radio);
    }
}
````
